```
acro_nyms=pd.read_csv('acronyms_dummy_code_data.csv',encoding = "ISO-8859-1")
```

Function that finds a unique values in a list.

```
def unique(list1): 
    counter1=0  
    ls=[]
    # insert the list to the set 
    list_set = set(list1) 
    # convert the set to the list 
    unique_list = (list(list_set)) 
    for x in unique_list: 
#         print(x)
        ls.append(x)
        counter1+=1
#     print(counter1)
    return ls
```

Extracts the intended code names to a list.
Extracts the intended definiton for the codes to a list.
Finds unique code values and saves to a seperate variable.
Assigns the dictionary keys with the unique code values and the values for the dictionary to a 0 for default.

Makes a for loop where uses the length of unfiltered appearence of the code names. If the key value in the dictionary is a 0, then overwrite with a list containing the paired definition value extracted from the definition list.  If there is a value that is not a 0, append the next definition to the list in the dictionary.

```
a_keys=acro_nyms.CD.to_list()
a_values=acro_nyms.NM.to_list()
the_u_keys=unique(a_keys)
acronym=dict.fromkeys(the_u_keys,0)
for key in range(len(a_keys)):
  if acronym[str(a_keys[key])] == 0:
    acronym[str(a_keys[key])] = [a_values[key]]
  else:
    acronym[str(a_keys[key])].append(a_values[key])
```
